```python
class student:
  def __init__(self,id,name,marks):
    self.id=id
    self.name=name
    self.marks=marks

  def display(self):
      print("ID ",self.id)
      print("Name ",self.name)
      print("Marks ",self.marks)

s1=student(112,"Naman",92)
s1.display()
```

```python
#inheritence

class parent:

  def test_parents(self) :
    print("this is my parent class")


```

```python
class child(parent):
  pass
```

```python
child_obj = child()
```

```python
child_obj.test_parents()
```

```python
class class1 :
  def test_class1(self):
    print("this is my class1")
```

```python
class class2(class1) :
  def test_class2(self) :
    print("this is my class2")
```

```python
class class3(class2) :
  def test_class3(self):
    print("this is my class3")
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class1()
```
